```python
import pandas as pd
import os
```

```python
myfiles = [x for x in os.listdir() if x.endswith("Datasheet.csv")]
myfiles
```

```python
fh = open(myfiles[0])
header = fh.read()
header = header.replace("RUN36", "Myositis")
header = header.replace("03-10-2019", "03-05-2021")
fh.close()
header = header.split("\n")[:20]
header
```

```python
def readDatasheet(f):
    cols = ['Sample_ID', 'Sample_Name', 'Sample_Plate', 'Sample_Well',
       'I7_Index_ID', 'index', 'I5_Index_ID', 'index2', 'Run', 'Nr',
       'Sample_Project', 'Species', 'Chain', 'Patient', 'Sample', 'UMI']
    df = pd.read_csv(f, skiprows=20)
    df = df[cols]
    df = df[df["Sample_Project"] == "Myositis"]
    return(df)

df = readDatasheet(myfiles[0])
for myfile in myfiles[1:]:
    df = pd.concat([df, readDatasheet(myfile)])
df
```

```python
# Write the header
fhOut = open("20210503-Myositis-datasheet.csv", "w")
print("\n".join(header), file=fhOut)
fhOut.close()

# Append the data to the previous file
df.to_csv("20210503-Myositis-datasheet.csv", mode='a', index=False)

print("Wrote 20210503-Myositis-datasheet.csv to disk")
```
