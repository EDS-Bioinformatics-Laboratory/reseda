```python
import pandas as pd
```

```python
df = pd.read_csv("20221027_MSrun050_pt-table-v2.csv", sep=";")
df.head()
```

```python
underscoreToDash = lambda x: x.replace("_", "-")
underscoreToDash("blah_something")
```

```python
df["Sample name"] = [x for x in map(underscoreToDash, df["Sample name"])]
df["Sample name"]
```

```python
df.to_csv("20221027_MSrun050_pt-table-v3.csv", sep=";", index=None)
```
