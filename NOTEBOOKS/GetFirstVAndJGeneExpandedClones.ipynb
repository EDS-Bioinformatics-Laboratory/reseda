```python
df = pd.read_csv('vjcdr3-clones-mut-BASTA-IGH_HUMAN-hecs.csv', sep="\t")
df.head()
```

```python
df.columns
```

```python
def getFirstVgene(gene_name):
    if "+" in gene_name:
        return(gene_name.split("+")[0])
    else:
        return(gene_name)

    
def getFirstJgene(gene_name):
    if "," in gene_name:
        return(gene_name.split(",")[0])
    else:
        return(gene_name)
```

```python
df['V'] = [x for x in map(getFirstVgene, df['V_sub'])]
df['J'] = [x for x in map(getFirstJgene, df['J_sub'])]
df.head()
```

```python
df.to_excel('vjcdr3-clones-mut-BASTA-IGH_HUMAN-hecs-VJ.xlsx')
print("Wrote vjcdr3-clones-mut-BASTA-IGH_HUMAN-hecs-VJ.xlsx to disk")
```
