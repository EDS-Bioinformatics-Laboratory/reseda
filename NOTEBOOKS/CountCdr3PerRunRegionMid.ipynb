```python
import pandas as pd
import os
```

```python
myfiles = [x for x in os.listdir() if x.endswith("-mids-CDR3-matches.csv")]
outfile = "lookup-CDR3-in-all-Roche.xlsx"
```

```python
cols = ['acc', 'CDR3pep', 'CDR3nuc', 'CDR3pep-short', 'CDR3nuc-short', 'CDR3', 'nt5', 'nt6', 'nt7',
       'nt8', 'nt9', 'nt10', 'nt11', 'nt12', 'nt13', 'nt14','CDR3_x', 'Samples']
df = pd.read_csv(myfiles[0])
df = df[cols]
df['file_name'] = myfiles[0]
for myfile in myfiles[1:]:
    df_tmp = pd.read_csv(myfile)
    df_tmp = df_tmp[cols]
    df_tmp['file_name'] = myfile
    df = pd.concat([df, df_tmp])
```

## Count the CDR3s per run, region, MID

```python
df_agg = df.groupby(['CDR3', 'Samples', 'file_name', 'nt10', 'nt14']).agg({'acc': len})
df_agg.head()
```

```python
df_agg.to_excel(outfile)
```
