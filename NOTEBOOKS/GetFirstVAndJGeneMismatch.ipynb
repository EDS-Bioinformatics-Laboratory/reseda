```python
df = pd.read_excel('basta-allinfo-mismatch-mut-nglyc-cdr3peplength.xlsx')
df.head()
```

```python
df.columns
```

```python
def getFirstVgene(gene_name):
    if "+" in gene_name:
        return(gene_name.split("+")[0])
    else:
        return(gene_name)

    
def getFirstJgene(gene_name):
    if "," in gene_name:
        return(gene_name.split(",")[0])
    else:
        return(gene_name)
```

```python
df['V'] = [x for x in map(getFirstVgene, df['V_sub_y'])]
df['J'] = [x for x in map(getFirstJgene, df['J_sub_y'])]
df.head()
```

```python
df.to_excel('basta-allinfo-mismatch-mut-nglyc-cdr3peplength-VJ.xlsx')
print("Wrote basta-allinfo-mismatch-mut-nglyc-cdr3peplength-VJ.xlsx to disk")
```
