```python
import os
import pandas as pd
```

```python
[x for x in os.listdir() if x.startswith("vjcdr3")]
```

```python
outfile = 'vjcdr3-clones-mut-BASTA-IGH_HUMAN-hecs.csv'
```

```python
df = pd.read_csv('vjcdr3-clones-mut-BASTA-IGH_HUMAN.csv', sep="\t")
df.head()
```

```python
print("Original length", len(df))
```

```python
# Check if the fraction UMIs sum up to 100 or 1. Answer: sums up to 1
df.groupby('Sample').agg({'UMIs.frac': sum})
```

```python
# Get expanded clones
df_hecs = df[df['UMIs.frac'] > 0.005]
df_hecs.head()
```

```python
print("Length df_hecs", len(df_hecs))
```

```python
# What is the total impact of HECs?
df_hecs.groupby('Sample').agg({'UMIs.frac': sum})
```

```python
# Export the HECs dataframe
df_hecs.to_csv(outfile, sep="\t")
print("Wrote", outfile, "to disk")
```
