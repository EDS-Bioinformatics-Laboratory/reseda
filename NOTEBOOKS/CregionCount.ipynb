```python
import os
import pandas as pd
```

```python
mydir = "../"
myfiles = [x for x in os.listdir(mydir) if x.endswith("-easy-import.txt")]
myfiles
```

```python
def readFile(f):
    sample, rest = f.split("_L001.assembled-")
    mid = rest.replace("-IGHC_CH12_human-easy-import.txt", "")
    
    df = pd.read_csv(f, sep="\t", header=None)
    df.columns = ["acc", "flag", "hit"]
    df["Sample"] = sample
    df["MID"] = mid
    df["filename"] = f
    return(df)
```

```python
df = readFile(mydir + myfiles[0])
for f in myfiles[1:]:
    df = pd.concat([df, readFile(mydir + f)])
df.tail()
```

```python
df_count = df.groupby(["Sample", "MID", "hit"]).agg({'acc': len})
df_count = df_count.reset_index()
df_count = df_count.sort_values(by=["Sample", "MID", "acc"], ascending=[True, True, False])
df_count
```

```python
df_count.to_excel("run42-RACE-C-count.xlsx")
print("Wrote run42-RACE-C-count.xlsx to disk")
```
