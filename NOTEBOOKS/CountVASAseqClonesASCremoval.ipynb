```python
import pandas as pd
```

```python
df_overview = pd.read_excel("overview-vasaseq-vjc-r2.xlsx", index_col=0, engine="openpyxl")
df_overview.head()
```

```python
df_noasc_clones = pd.read_excel("VASAseq-noASC-CDR3-IGH-clones.xlsx", index_col=0, engine="openpyxl")
df_noasc_clones.head()
```

```python
df_count_igh = df_noasc_clones.groupby("Sample").agg({"cdr3pep": "nunique", "freq": sum})
df_count_igh = df_count_igh.reset_index()
df_count_igh = df_count_igh.rename(columns={"Sample": "barcode","cdr3pep": "unique_cdr3pep", "freq": "nr_cdr3_igh_noASC"})
df_count_igh.head()
```

```python
df_overview_new = df_overview.merge(df_count_igh, how="left", on="barcode")
df_overview_new = df_overview_new.fillna(0)
df_overview_new.head()
```

```python
df_overview_new.to_excel("overview-vasaseq-vjc-r2-noASC.xlsx")
```
