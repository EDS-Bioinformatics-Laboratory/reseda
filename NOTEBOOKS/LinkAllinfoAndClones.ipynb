```python
import pandas as pd
import os
```

```python
clonesFiles = [x for x in os.listdir() if x.endswith("-clones-mut-sites-reassigned.csv")]
```

```python
df_basta = pd.read_excel("basta-allinfo-long.xlsx")
df_basta.head()
```

```python
df_basta['cdr3pep_length'] = df_basta["cdr3pep"].str.len()
df_basta.head()
```

```python
df_clones = pd.read_csv(clonesFiles[0], sep="\t")
sample_name, rest = clonesFiles[0].split("_L001")
df_clones["Sample"] = sample_name
df_clones.head()
```

```python
for myfile in clonesFiles[1:]:
    df_tmp = pd.read_csv(myfile, sep="\t")
    sample_name, rest = myfile.split("_L001")
    df_tmp["Sample"] = sample_name
    df_clones = pd.concat([df_clones, df_tmp])
```

```python
df_combined = pd.merge(df_basta, df_clones, on=['Sample', 'cdr3pep'])
df_combined.head()
```

```python
df_combined.to_excel("basta-allinfo-long-mut-nglyc-cdr3peplength.xlsx")
print("Wrote basta-allinfo-long-mut-nglyc-cdr3peplength.xlsx to disk")
```
