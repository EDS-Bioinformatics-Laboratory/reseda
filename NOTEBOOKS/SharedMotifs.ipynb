```python
truncate = lambda x: str(x)[3:-7-5]
patient_col = "patient"
sample_col = "acc"
seq_col = "cdr3nuc"
df['cdr3_truncated'] = [x for x in map(truncate, df[seq_col])]  # LET OP: nog bepalen hoeveel precies eruit moet
seq_col = 'cdr3_truncated'
cols = [patient_col, sample_col, seq_col]
df[cols].head()
```

```python
df_uniq = df[cols].groupby(cols).count()
df_uniq = df_uniq.reset_index()
df_uniq.columns = ["Patient", "Sample", "Sequence"]
df_uniq.head()
```

## Count motifs of a certain length

```python
def get_kmers(sequence, kmer_length):
    kmers = list()
    for i in range(0, len(sequence) - kmer_length + 1):
        kmers.append(sequence[i:i + kmer_length])
    return(kmers)
print("ABCDEFGH", get_kmers("ABCDEFGH", 3))
print("AB", get_kmers("AB", 3))
print("ABCDEF", get_kmers("ABCDEF", 3))
print("ABCDEF", get_kmers("ABCDEF", 1))
```

```python
cols_kmers = ['Patient', 'Sample', 'kmer']
df_kmers = pd.DataFrame(columns=cols_kmers)
for index, row in df_uniq.iterrows():
    kmers = get_kmers(row['Sequence'], 3)
    df_tmp = pd.DataFrame({'Patient': row['Patient'], 'Sample': row['Sample'], 'kmer': kmers})
    df_kmers = pd.concat([df_kmers, df_tmp])
df_kmers = df_kmers.reset_index()
df_kmers = df_kmers[cols_kmers]
```

```python
concat_values = lambda x: ','.join(list(set(x)))
df_kmer_count = df_kmers.groupby('kmer').agg({'Patient': ['nunique', concat_values], 'Sample': ['nunique', concat_values]})
df_kmer_count = df_kmer_count.sort_values(by=[('Patient','nunique'), ('Sample', 'nunique')], ascending=False)
df_kmer_count
```

## Check if the above kmers are part of the V, D or J sequence

```python
from Bio import SeqIO
```

```python
refAA = dict()

fhIn = open("/home/barbera/git/reseda/reference/human_IGHJ_AA.fasta")
for record in SeqIO.parse(fhIn, "fasta"):
    myId = str(record.id).split("|")[1]
    mySeq = str(record.seq)
    refAA[myId] = refAA.get(myId, "") + mySeq
fhIn.close()
```

```python
# Translate the D gene in 3 reading frames and add it to the refAA dictionary
fhIn = open("/home/barbera/git/reseda/reference/IGHD_human.fasta")
for record in SeqIO.parse(fhIn, "fasta"):
    myId = str(record.id).split("|")[1] + "_1"
    mySeq = str(record.seq.translate())
    refAA[myId] = refAA.get(myId, "") + mySeq
    
    myId = str(record.id).split("|")[1] + "_2"
    mySeq = str(record.seq[1:].translate())
    refAA[myId] = refAA.get(myId, "") + mySeq
    
    myId = str(record.id).split("|")[1] + "_3"
    mySeq = str(record.seq[2:].translate())
    refAA[myId] = refAA.get(myId, "") + mySeq
fhIn.close()
```

```python
hits = dict()
for kmer in df_kmer_count.index:
    hits[kmer] = list()
    for myId, seq in refAA.items():
        if kmer in seq:
            print(kmer, myId, seq)
            hits[kmer].append(myId)
```

```python
# Concatenate the hits (comma separated)
hits_kmer = list()
hits_hit = list()
for key, value in hits.items():
    hits_kmer.append(key)
    hits_hit.append(",".join(value))
```

```python
df_hits = pd.DataFrame({'kmer': hits_kmer, 'hits': hits_hit})
df_hits.head()
```

```python
df_combined = pd.merge(df_kmer_count, df_hits, how='left', on='kmer')
```
