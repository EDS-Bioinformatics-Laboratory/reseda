```python
import pandas as pd
import os
```

```python
fh = open("SAMPLES-fastq")
samples = fh.read().split("\n")
fh.close()
samples[:10]
```

```python
f = lambda x: x.split("/")[-1]
samples = [x for x in map(f, samples)]
samples[:10]
```

```python
for sample in samples:
    cdr3_file = sample + "-IGH_HUMAN-CDR3-matches.csv"
    mid_file = sample + "-mids.csv"
    out_file = sample + "-mids-CDR3-matches.csv"
    
    # Try reading the files, in case of error, go to the next sample
    try:
        df_cdr3 = pd.read_csv(cdr3_file, sep=",")
        df_mid = pd.read_csv(mid_file, sep="\t", skiprows=1, header=None)
        df_mid.columns = ['acc','nt5','nt6','nt7','nt8','nt9','nt10','nt11','nt12','nt13','nt14']
    except:
        continue
        
    if df_mid['nt5'][0].startswith('gact') or df_mid['nt5'][0].startswith('tcag'):
        df_mid = df_mid.replace({'^gact': '', '^tcag': ''}, regex=True)
        
    df = pd.merge(df_cdr3, df_mid, how='inner', on='acc')
    df.to_csv(out_file)
    print("Wrote", out_file, "to disk")
```
