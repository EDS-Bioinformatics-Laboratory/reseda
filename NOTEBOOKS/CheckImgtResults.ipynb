```python
import pandas as pd
```

```python
df_bcr = pd.read_csv("./IMGT-BCR/vquest_airr.tsv", sep="\t", index_col=False)
df_bcr.head()
```

```python
df_tcr = pd.read_csv("./IMGT-TCR/vquest_airr.tsv", sep="\t", index_col=False)
df_tcr.head()
```

```python
def data_to_cols(x):
    barcode, contig, size, read, cov = x.split("|")
    size = int(size.replace("size", ""))
    read = int(read.replace("read", ""))
    cov = float(cov.replace("cov", ""))
    return(barcode, contig, size, read, cov)

data_to_cols("CCAGATGA|contig1|size95|read4|cov3.57")
```

```python
df_bcr[["barcode", "contig", "contig_size", "contig_reads", "contig_coverage"]] = [x for x in map(data_to_cols, df_bcr["sequence_id"])]
df_bcr.head()
```

```python
df_tcr[["barcode", "contig", "contig_size", "contig_reads", "contig_coverage"]] = [x for x in map(data_to_cols, df_tcr["sequence_id"])]
df_tcr.head()
```

```python
df_bcr.to_excel("./IMGT-BCR/vquest-bcr-contig-info.xlsx")
df_tcr.to_excel("./IMGT-TCR/vquest-tcr-contig-info.xlsx")
```

```python
# how many BCR CDR3s were found?
df_bcr_junctions = df_bcr.groupby(["barcode", "locus", "junction_aa"]).agg({"sequence_id": len}).sort_values("sequence_id", ascending=False)
df_bcr_junctions
```

```python
# how many TCR CDR3s were found?
df_tcr_junctions = df_tcr.groupby(["barcode", "locus", "junction_aa"]).agg({"sequence_id": len}).sort_values("sequence_id", ascending=False)
df_tcr_junctions
```

```python
df_bcr_junctions.to_excel("./IMGT-BCR/brc-junctions.xlsx")
```

```python
df_tcr_junctions.to_excel("./IMGT-TCR/tcr-junctions.xlsx")
```

```python
# Count barcodes with more than one locus
df = pd.concat([df_bcr_junctions, df_tcr_junctions])
df = df.reset_index()
df.head()
```

```python
concatenate = lambda x: ",".join(list(set(x)))
df_count_loci = df.groupby("barcode").agg({"locus": ["nunique", concatenate], "junction_aa": ["nunique", concatenate]}).sort_values(("junction_aa", "nunique"), ascending=False)
df_count_loci
```

```python
df_count_loci.to_excel("./IMGT-BCR/count-loci-bcr-tcr.xlsx")
```
