```python
import pandas as pd
import numpy
```

```python
df = pd.read_csv("PR3-final-folder.csv")
df.head()
```

```python
df = df.fillna(0)
```

```python
pr3_list = list(df['Total list PR3 motifs'])
pr3_list
```

```python
pr3_list = [x for x in pr3_list if x != 0]
pr3_list
```

```python
found_list = list()
found_motifs = list()
for i in range(len(df)):
    found = False
    tmp_motif_list = list()
    for motif in pr3_list:
        if motif in df["Unnamed: 0"][i]:  # CDR3pep
            found = True
            tmp_motif_list.append(motif)
    found_motifs.append(",".join(tmp_motif_list))
    found_list.append(found)
df["PR3-present"] = found_list
df["PR3-motifs-found"] = found_motifs
df.head()
```

```python
df.to_csv("PR3-final-folder-motifsearch.csv")
```
