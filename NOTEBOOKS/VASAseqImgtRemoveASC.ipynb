```python
import os
import pandas as pd
```

```python
# Read well positions and barcodes
df_wells = pd.read_excel("/home/barbera/Downloads/VASAseq/celseq2_primers.xlsx", engine="openpyxl")
df_wells = df_wells.rename(columns={'Well Position': 'well', 'Cell Barcode': 'barcode'})
cols = ['barcode', 'well']
df_wells = df_wells[cols]
df_wells.head()
```

```python
# Read IMGT tables
df_bcr = pd.read_excel("IMGT-BCR/vquest-bcr-contig-info.xlsx", engine="openpyxl", index_col=0)
df_bcr.head()
```

```python
df_tcr = pd.read_excel("IMGT-TCR/vquest-tcr-contig-info.xlsx", engine="openpyxl", index_col=0)
df_tcr.head()
```

```python
df = pd.concat([df_bcr, df_tcr])
df.head()
```

```python
# link well id to IMGT table
df = pd.merge(df, df_wells, on="barcode")
print("Entries:", len(df))
df.head()
```

## Fetch CDR3s belonging to antibody secreting cells and remove from table

```python
# ASC B cells: K1-24, D19-21
asc = list()
for i in range(1,25):
    asc.append("K" + str(i))
for i in range(19,22):
    asc.append("D" + str(i))
print(asc)
```

```python
# Get ASC clones
df_asc = df[df['well'].isin(asc)]
print("Entries:", len(df_asc))
df_asc.head()
```

```python
# Remove ASC clones from original clones table
df = df[~df['junction_aa'].isin(df_asc['junction_aa'])]
print("Entries:", len(df))
df.head()
```

```python
df_summary = df.groupby(["barcode", "well", "locus", "junction_aa"]).agg({'sequence_id': len}).sort_values(["well", "locus"])
df_summary
```
