```python
import os
import pandas as pd
```

```python
myfiles = [x for x in os.listdir() if x.endswith("-IGH_HUMAN-report.txt")]
myfiles.sort()
myfiles
```

```python
def read_file(myfile):
    sample = myfile.split("_L001")[0]
    df_tmp = pd.read_csv(myfile, skiprows=1, sep="\t", header=None)
    df_tmp["Sample"] = sample
    df_tmp = df_tmp.rename(columns={0: 'Description', 1: 'Reads', 2: 'Percentage'})
    return(df_tmp)

df = read_file(myfiles[0])
for myfile in myfiles[1:]:
    df = pd.concat([df, read_file(myfile)])
    
df = df.reset_index(drop=True)
df
```
