```python
import pandas as pd
```

```python
clonesFile = "AMC-TC-x001-HL-clones.csv"
donorFile = "seurat_coupling_mRNA_HTO_VDJ.raw.txt"
```

```python
df_clones = pd.read_csv(clonesFile)
df_clones.head()
```

```python
df_donors = pd.read_csv(donorFile, sep="\t")
df_donors.head()
```

```python
df = pd.merge(df_clones, df_donors, how="inner", on="barcode")
df.head()
```

## Check what the overlap is

```python
barcodes_clones = set(df_clones["barcode"])
barcodes_donors = set(df_donors["barcode"])
```

```python
print("barcodes in clones:", len(barcodes_clones))
print("barcodes in donors:", len(barcodes_donors))
print("clones diff donors:", len(barcodes_clones.difference(barcodes_donors)))
print("donors diff clones:", len(barcodes_donors.difference(barcodes_clones)))
print("donors intersection clones:", len(barcodes_donors.intersection(barcodes_clones)))
print("donors union clones:", len(barcodes_donors.union(barcodes_clones)))
```

## Is number of chains consistent with the singlet/doublet annotation?

```python
df_clones_single_chains = df[(df["cdr3_heavy_nunique"] == 1) & (df["cdr3_light_nunique"] == 1)]
df_clones_single_chains.groupby("HTO_classification.global").agg({'barcode': "count"})
```

```python
df_clones_multiple_chains = df[(df["cdr3_heavy_nunique"] > 1) | (df["cdr3_light_nunique"] > 1)]
df_clones_multiple_chains.groupby("HTO_classification.global").agg({'barcode': "count"})
```
