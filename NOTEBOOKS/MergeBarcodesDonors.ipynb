```python
import pandas as pd
```

```python
clonesFile = "/home/barbera/Downloads/runXXX-20210107-SC-Tom/processing/20211102-reseda/20211103-merge-heavy-light/AMC-TC-x001-HL-clones-mut.csv"
donorFile = "/home/barbera/Downloads/runXXX-20210107-SC-Tom/aldo-perry/coupling_Seurat_RESEDA_Info.raw.05022021_manually_edited.xlsx"
```

```python
outFile = "AMC-TC-x001-HL-clones-mut-donors.csv"
```

```python
df_clones = pd.read_csv(clonesFile)
print("Entries:", len(df_clones))
df_clones.head()
```

```python
df_donors = pd.read_excel(donorFile, sheet_name=1, skiprows=2)
print("Entries:", len(df_donors))
df_donors.head()
```

```python
# Only keep the Seurat data, and remove the old Reseda data
cols = ['barcode', 'nCount_RNA', 'nFeature_RNA', 'percent.mito',
       'n.exp.hkgenes', 'S.Score', 'G2M.Score', 'Phase', 'nCount_HTO',
       'nFeature_HTO', 'HTO_maxID', 'HTO_secondID', 'HTO_margin',
       'HTO_classification', 'HTO_classification.global', 'hash.ID',
       'MULTI_ID', 'MULTI_classification', 'nCount_ADT', 'nFeature_ADT',
       'CD27_counts', 'IgD_counts', 'CD27_normCounts_CLR',
       'IgD_normCounts_CLR', 'IGH', 'cdr3_aa1', 'cdr3_nt1', 'IGLC', 'cdr3_aa2',
       'cdr3_nt2', 'CTgene', 'CTnt', 'CTaa', 'CTstrict', 'raw_clonotype_id',
       'raw_consensus_id', 'frequency_10X_cloneFile',
       'proportion_10X_cloneFile', 'reads_heavy_chain', 'reads_light_chain',
       'umis_heavy_chain', 'umis_light_chain', 'frequency_recalculated',
       'proportion_recalculated']
df_donors = df_donors[cols]
df_donors.head()
```

```python
df = pd.merge(df_donors, df_clones, how="inner", on="barcode")
print("Entries:", len(df))
df.head()
```

```python
df.to_csv(outFile)
print("Wrote", outFile, "to disk")
```

## Check what the overlap is

```python
barcodes_clones = set(df_clones["barcode"])
barcodes_donors = set(df_donors["barcode"])
```

```python
print("barcodes in clones:", len(barcodes_clones))
print("barcodes in donors:", len(barcodes_donors))
print("clones diff donors:", len(barcodes_clones.difference(barcodes_donors)))
print("donors diff clones:", len(barcodes_donors.difference(barcodes_clones)))
print("donors intersection clones:", len(barcodes_donors.intersection(barcodes_clones)))
print("donors union clones:", len(barcodes_donors.union(barcodes_clones)))
```

## Is number of chains consistent with the singlet/doublet annotation?

```python
df_clones_single_chains = df[(df["cdr3_heavy_nunique"] == 1) & (df["cdr3_light_nunique"] == 1)]
df_clones_single_chains.groupby("HTO_classification.global").agg({'barcode': "count"})
```

```python
df_clones_multiple_chains = df[(df["cdr3_heavy_nunique"] > 1) | (df["cdr3_light_nunique"] > 1)]
df_clones_multiple_chains.groupby("HTO_classification.global").agg({'barcode': "count"})
```
