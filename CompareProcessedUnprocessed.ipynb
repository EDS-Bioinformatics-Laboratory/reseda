```python
import pandas as pd
```

```python
df_orig = pd.read_csv("20240227-DataSheet-RUN54.csv", skiprows=20)
df_orig.head()
```

```python
# Preparation: ls /mnt/immunogenomics/RUNS/$RUN/Processing/$OUTDIR/Results/final/*.rr.clones_subs.csv > list-processed.txt
df_processed = pd.read_csv("list-processed.txt", header=None)
df_processed.columns = ['filename']
df_processed.head()
```

```python
get_sample = lambda x: x.split("/")[-1].split('_')[0]
df_processed['Sample_Name'] = [x for x in map(get_sample, df_processed['filename'])]
df_processed.head()
```

```python
df = pd.merge(df_orig, df_processed, how='left', on='Sample_Name')
df
```

```python
# Get list of unprocessed samples
df = df[df['filename'].isna()]
len(df)
```

```python
# Preparation: ls /mnt/immunogenomics/RUNS/$RUN/Data/NameOfDataset_1/Raw/*.fastq.gz > list-fastq.txt
df_fastq = pd.read_csv("list-fastq.txt", header=None)
df_fastq.columns = ['fastq']
df_fastq.head()
```

```python
get_sample = lambda x: x.split("/")[-1].split('_')[0]
df_fastq['Sample_Name'] = [x for x in map(get_sample, df_fastq['fastq'])]
df_fastq.head()
```

```python
# Combined list unprocessed samples with fastq paths
df = pd.merge(df, df_fastq, how='left', on='Sample_Name')
df.head()
```

```python
len(df)
```

```python
# Check which reference should be used
df.groupby(['Species', 'Chain', 'UMI', 'Constant', 'RACE']).agg({'Sample_Name': 'nunique'})
```

```python
df_human_TCRb = df[(df['Species'] == 'human') & (df['Chain'] == "TCRb")]
df_human_BCRh_UMI = df[(df['Species'] == 'human') & (df['Chain'] == "BCRh") & (df['UMI'] == "yes") & (df['RACE'] == "no")]
df_human_BCRh_RACE_noC = df[(df['Species'] == 'human') & (df['Chain'] == "BCRh") & (df['RACE'] == "yes") & (df['Constant'] == "no")]
df_human_BCRh_RACE_withC = df[(df['Species'] == 'human') & (df['Chain'] == "BCRh") & (df['RACE'] == "yes") & (df['Constant'] == "yes")]
df_mouse_BCRk = df[(df['Species'] == 'mouse') & (df['Chain'] == "BCRk")]
```

```python
df_human_TCRb = df_human_TCRb['fastq']
df_human_BCRh_UMI = df_human_BCRh_UMI['fastq']
df_human_BCRh_RACE_noC = df_human_BCRh_RACE_noC['fastq']
df_human_BCRh_RACE_withC = df_human_BCRh_RACE_withC['fastq']
df_mouse_BCRk = df_mouse_BCRk['fastq']
```

```python
df_human_TCRb.to_csv("SAMPLES-human-TCRb", header=False, index=False)
df_human_BCRh_UMI.to_csv("SAMPLES-human-BCRh-UMI", header=False, index=False)
df_human_BCRh_RACE_noC.to_csv("SAMPLES-human-BCRh-RACE-noC", header=False, index=False)
df_human_BCRh_RACE_withC.to_csv("SAMPLES-human-BCRh-RACE-withC", header=False, index=False)
#df_mouse_BCRk.to_csv("SAMPLES-mouse-BCRk", header=False, index=False)
```
