```python
import pandas as pd
import os
```

## Read sample information

```python
df = pd.read_excel("20160201_RUN259_MIDlijst.xls", header=1)
df.head()
```

```python
region = 2
prefix = 'run259-JV3AIYM02-'
```

```python
selection = df.loc[df['SAMPLE.ID'].str.startswith('JBI')]
selection = selection[selection['REGION'] == region]
selection.head()
```

```python
d = dict()
cols = ['MID.SEQ', 'SAMPLE.ID']
for mid, sampleId in selection[cols].to_dict(orient='split')['data']:
    d[mid] = sampleId
d
```

## Rename fastq files

```python
myfiles = list()
for f in os.listdir('split'):
    if f.startswith(prefix) and f.endswith('.fastq.gz'):
        myfiles.append(f)
print(myfiles)
```

```python
s = 1
for f in myfiles:
    mid = f.replace(prefix, '')
    mid = mid.replace('.fastq.gz', '')
    sampleId = d.get(mid, 'unknown')
    new_f = sampleId + "-" + f
    new_f = new_f.replace('.fastq.gz', '_S' + str(s) + '_L001.assembled.fastq.gz')
    print("mv", f, new_f)
    s = s + 1
```
