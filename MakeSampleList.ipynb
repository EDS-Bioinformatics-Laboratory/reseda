```python
samples_run35 = pd.read_excel('Miseq-run035-murine-BCRH-samples.xlsx')
samples_run35["SampleID"] = samples_run35["Sample name"] + "_" + samples_run35["Samples order"]
samples_run35.head()
```

```python
samples_run40 = pd.read_excel('Miseq-run040-murine-BCRH-samples.xlsx')
samples_run40["SampleID"] = samples_run40["Sample name"] + "_" + samples_run40["Samples order"]
samples_run40.head()
```

```python
mydir = "/mnt/immunogenomics/RUNS/run35-20190609-miseq/data"
files_run35 = list()
for myfile in os.listdir(mydir):
    if myfile.split("_L001")[0] in list(samples_run35["SampleID"]):
        files_run35.append(mydir + "/" + myfile)
files_run35[:10]
```

```python
mydir = "/mnt/immunogenomics/RUNS/run40-20201116-miseq/data"
files_run40 = list()
for myfile in os.listdir(mydir):
    if myfile.split("_L001")[0] in list(samples_run40["SampleID"]):
        files_run40.append(mydir + "/" + myfile)
files_run40[:10]
```

```python
fhOut = open("SAMPLES-run35-mouse", "w")
fhOut2 = open("SAMPLES-run40-mouse", "w")
for myfile in files_run35:
    print(myfile, file=fhOut)
for myfile in files_run40:
    print(myfile, file=fhOut2)
fhOut.close()
fhOut2.close()
print("Wrote SAMPLES-run35-mouse to disk")
print("Wrote SAMPLES-run40-mouse to disk")
```
