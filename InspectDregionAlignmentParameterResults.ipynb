```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
mydir = "DREGION-TRB-SAMPLES/"
myfiles = [mydir + x for x in os.listdir(mydir) if x.endswith("-summary.csv")]
myfiles[:10]
```

```python
df = pd.read_csv(myfiles[0])
df.head()
```

```python
for myfile in myfiles[1:]:
    df = pd.concat([df, pd.read_csv(myfile)])
df.tail()
```

```python
df_score = df[df['align_info'] == "alignLength"]
df_score.head()
```

```python
plt.hist(df_score['mean'])
plt.show()
```

```python
df_score[df_score['mean'] == df_score['mean'].min()]
```

```python
df_score[df_score['mean'] == df_score['mean'].max()]
```

```python
plt.scatter(df_score['score_gap'], df_score['mean'])
plt.xlabel('score_gap')
plt.ylabel('mean alignLength')
plt.show()
```

```python
plt.scatter(df_score['score_match'], df_score['mean'])
plt.xlabel('score_match')
plt.ylabel('mean alignLength')
plt.show()
```

```python
plt.scatter(df_score['score_mismatch'], df_score['mean'])
plt.xlabel('score_mismatch')
plt.ylabel('mean alignLength')
plt.show()
```
