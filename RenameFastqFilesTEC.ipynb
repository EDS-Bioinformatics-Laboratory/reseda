```python
import pandas as pd
import os
```

```python
df_pt = pd.read_excel("project-20230912-TEC-TCRb/Data/NameOfDataset_1/Meta/TEC-samples.xlsx")
df_pt.head()
```

```python
myfiles = [x for x in os.listdir("split") if x.endswith(".fastq.gz")]
print(myfiles)
```

```python
runs = list()
regions = list()
mids = list()
for myfile in myfiles:
    run, region, machine, mid = myfile.split("-")
    region = int(region.replace("r", ""))
    mid = mid.split(".")[0]
    runs.append(run)
    regions.append(region)
    mids.append(mid)
df_files = pd.DataFrame({'RUN': runs, 'region': regions, "MID": mids, 'filename': myfiles})
df_files.head()
```

```python
df = pd.merge(df_pt, df_files, on=['RUN', 'region', 'MID'])
df
```

```python
df["filename-new"] = df["Sample_Name"].str.replace(" ", "-") + "_S1_L001.assembled.fastq.gz"
df
```

```python
for i, row in df.iterrows():
    print("mv", row['filename'], row['filename-new'])
```
